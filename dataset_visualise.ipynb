## Visualising Datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = r"C:\Users\raksh\handwritten-digits-recogniser\MNIST_CSV\mnist_train.csv" 
data = pd.read_csv(file_path)

print(data.head())

labels = data.iloc[:, 0]  # first column is the label
features = data.iloc[:, 1:]  # Remaining columns are pixel values

classes = labels.unique()
plt.figure(figsize=(12, 8))

for i, cls in enumerate(classes):
    sample_index = labels[labels == cls].index[0]
    sample_image = np.array(features.iloc[sample_index]).reshape(28, 28)
    plt.subplot(2, 5, i + 1)
    plt.imshow(sample_image, cmap="gray")
    plt.title(f"Class: {cls}")
    plt.axis("off")

plt.tight_layout()
plt.show()

num_columns = len(data.columns)
print(f"Number of columns: {num_columns}")
```
